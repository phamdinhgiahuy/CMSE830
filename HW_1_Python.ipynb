üìù **Answer to Problem 2**
* cardinal number: is the count by which we can get a quantitative grasp of the thing it represents. It is a numeric value that we can do calculation on.
* ordinal number: there is an inherent order or ranking for this type of number, it represents a clear organization of categories showing which comes first and which comes next. 
* nominal number: is like a name or a lable we give to an object, the purpose is just to classify or distinguish between them, there is no order or ranking of any sort.

    üëâ Depending on the interest and the inherent nature of a data science problems we can choose how to record, categorize and represent our data in a meaningful way while making sure that we do not take on unnecessary computational resource. For example if we just want to distinguish 3 different type of apples, there no need to assign a cardinal or ordinal number for each type to indicate the rank or the interval, but if we want to seperate these apple into different sizes (small, medium, large) then we have to assign ordinal numbers to them.


‚úçüèº **Problem 3**: Write a paragraph about John W. Tukey. You can use the class sides, but also Wikipedia or any other source. Why is he important to data science? Open a markdown cell below this one to put your discussion.

üìù **Answer to Problem 3**

John Wilder Tukey was the pioneer who laid several key foundations of the Data Science as widely known today. His concept of Exploratory Data Analysis (EDA) revolutionized data science, it emphasizes the importance of visualizing and understanding data to explore and gain insights from them, which in turn reveals meaningful patterns and outliers. In many ways, EDA was a precursor to Data Science so it is fair to consider John Wilder Tukey the father of modern Data Science. Tukey has soon realized and harnessed the power of computer science especially computer graphics to make studying multivariate data much easier. He was also the creator of many statistical methods and procedures that are widely used in data analysis, including the Fast Fourier Transform (FFT) algorithm, which is a key concept in signal processing and data compression. His legacy is a long lasting one and the path that he trod has led to unimaginable branches and applications of modern day Data Science.

üìù **Answer to Problem 6**

* a tuple is a container of flexible types of elements, these elements maintain the order they are created with. A tuple is denoted by the enclosing `()`
* a list is a container of flexible types of elements, these elements maintain the order they are created with. A list is denoted by the enclosing `[]`
* a array is a typical data structure in the `NumPy` library. It can only store one type of data at a time.
* the key differences and similarities between these three are:
    * tuple and list both contain flexible typing, ordered elements but tuple is immutable while list is mutable, in other words, tuple is like an unchangable list and list is a flexible, easy to modify data container.
    * The key difference of array from tuple and list is that all the elements in an array have to be of the same type while list and tuple simutaneously accept any type of data as their elements. 
* We need all three of these because each specific problems may call for the usage of one of these three. Each of these three has their own advantage and disavantage, for example while list and tuple are tolerent towards data types they are much less efficient in terms of computation compared to array, especially on large datasets. Or the ability to be modify with ease for list can be both its strength and shortcoming, it makes list versatile but less reliable than tuple so in cases where we need our data to be fixed and unchangeable we should use tuple, for like a dictionary key.
